### 첨도와 왜도
- 첨도
    - 데이터 분포의 꼬리의 두꺼움을 나타내는 지표
    - 양의 첨도 : 분포의 꼬리가 두껍고, 중심에 더 많은 데이터가 몰려 있음
    - 0 첨도 : 정규 분포와 유사한 분포를 가짐
    - 음의 첨도 : 분포의 꼬리가 얇고, 데이터가 중심에서 퍼져있음
- 왜도
    - 데이터 분포의 비대칭성을 나타내는 지표
    - 양의 왜도 : 오른쪽 꼬리가 길고, 데이터가 왼쪽으로 치우쳐 있음
    - 0 왜도 : 데이터가 대칭적으로 분포
    - 음의 왜도 : 왼쪽 꼬리가 길고, 데이터가 오른쪽으로 치우쳐 있음

```python
import numpy as np
import pandas as pd

df = pd.read_csv("dataset/주택 가격/train.csv")
df.head()
```

```python
# 수치형 데이터만 사용
df = df.select_dtypes(exclude='object')
df.head()
```

```python
# 중앙값으로 결측치 채우기
df = df.fillna(value=df.median())
```

```python
# 1. 왜도
skewness = df.skew()
ans = skewness.idxmax()
print(ans, skewness.max())
```

```python
# 2. 첨도
kurtosis = df.kurt()
ans = kurtosis.idxmax()
print(ans, kurtosis.max())
```
